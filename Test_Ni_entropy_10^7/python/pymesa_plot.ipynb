```python
from scipy.interpolate import interp1d

def create_interpolation_function(original_m, original_quantity):
    interpolation_function = interp1d(original_m, original_quantity, kind='linear', fill_value="extrapolate")
    return interpolation_function
```

```python
new_m = np.linspace(0., 0.64, 3000)  #New mass coordinate range
```

```python
internal_energy_donor=create_interpolation_function(relaxed_donor_remnant.q*0.64,relaxed_donor_remnant.energy)
internal_energy_donor_heated=create_interpolation_function(evolved_donor_remnant_mesatemp.q*0.64,evolved_donor_remnant_mesatemp.energy)
internal_energy_wd_inital=create_interpolation_function(evolved_mesa_initial.q*0.64,evolved_mesa_initial.energy)
internal_energy_wd_final=create_interpolation_function(evolved_mesa.q*0.64,evolved_mesa.energy)
```

```python
#New energies
ie_donor = internal_energy_donor(new_m)
ie_donor_heated = internal_energy_donor_heated(new_m)
ie_wd=internal_energy_wd_inital(new_m)
ie_wd_heated=internal_energy_wd_final(new_m)
```

```python
delta_e_arepo=ie_donor_heated-ie_donor
delta_e_mesa=ie_wd_heated-ie_wd
```

```python
pressure_donor_func=create_interpolation_function(relaxed_donor_remnant.q*0.64,np.power(10,relaxed_donor_remnant.logP))
pressure_donor_heated_func=create_interpolation_function(evolved_donor_remnant_mesatemp.q*0.64,np.power(10,evolved_donor_remnant_mesatemp.logP))
pressure_wd_inital_func=create_interpolation_function(evolved_mesa_initial.q*0.64,np.power(10,evolved_mesa_initial.logP))
pressure_wd_final_func=create_interpolation_function(evolved_mesa.q*0.64,np.power(10,evolved_mesa.logP))
```

```python
#New Pressures
pressure_donor = pressure_donor_func(new_m)
pressure_donor_heated = pressure_donor_heated_func(new_m)
pressure_wd=pressure_wd_inital_func(new_m)
pressure_wd_heated=pressure_wd_final_func(new_m)
```

```python
density_donor_func=create_interpolation_function(relaxed_donor_remnant.q*0.64,np.power(10,relaxed_donor_remnant.logRho))
density_donor_heated_func=create_interpolation_function(evolved_donor_remnant_mesatemp.q*0.64,np.power(10,evolved_donor_remnant_mesatemp.logRho))
density_wd_inital_func=create_interpolation_function(evolved_mesa_initial.q*0.64,np.power(10,evolved_mesa_initial.logRho))
density_wd_final_func=create_interpolation_function(evolved_mesa.q*0.64,np.power(10,evolved_mesa.logRho))
```

```python
#New Densities
density_donor = density_donor_func(new_m)
density_donor_heated = density_donor_heated_func(new_m)
density_wd=density_wd_inital_func(new_m)
density_wd_heated=density_wd_final_func(new_m)
```

```python
pdv_arepo=-(pressure_donor*(density_donor_heated-density_donor))/(density_donor**2)
pdv_mesa=-(pressure_wd*(density_wd_heated-density_wd))/(density_wd**2)
```

```python
temp_donor_func=create_interpolation_function(relaxed_donor_remnant.q*0.64,np.power(10,relaxed_donor_remnant.logT))
temp_wd_inital_func=create_interpolation_function(evolved_mesa_initial.q*0.64,np.power(10,evolved_mesa_initial.logT))
```

```python
#New Temperatures
temp_donor = temp_donor_func(new_m)
temp_wd=temp_wd_inital_func(new_m)
```

```python
entropy_donor_func=create_interpolation_function(relaxed_donor_remnant.q*0.64,np.power(10,relaxed_donor_remnant.logS))
entropy_donor_heated_func=create_interpolation_function(evolved_donor_remnant_mesatemp.q*0.64,np.power(10,evolved_donor_remnant_mesatemp.logS))
entropy_wd_inital_func=create_interpolation_function(evolved_mesa_initial.q*0.64,np.power(10,evolved_mesa_initial.logS))
entropy_wd_final_func=create_interpolation_function(evolved_mesa.q*0.64,np.power(10,evolved_mesa.logS))
```

```python
#New Densities
entropy_donor = entropy_donor_func(new_m)
entropy_donor_heated = entropy_donor_heated_func(new_m)
entropy_wd=entropy_wd_inital_func(new_m)
entropy_wd_heated=entropy_wd_final_func(new_m)
```

```python
tds_arepo=(temp_donor*(entropy_donor_heated-entropy_donor))
tds_mesa=(temp_wd*(entropy_wd_heated-entropy_wd))
```
