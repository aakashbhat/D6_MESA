```python
sigma=5.670e-5
heightsquare=np.square(relaxed_donor_remnant.pressure_scale_height*6.96e10)
diffusion=16*sigma*np.power(relaxed_donor_remnant.temperature,3)/(3*relaxed_donor_remnant.opacity*np.square(relaxed_donor_remnant.rho)*relaxed_donor_remnant.cp)
```
